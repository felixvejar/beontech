```python
destinations_ewr = nyc_flights[nyc_flights['origin'] == 'EWR']

destinations_ewr.head()
```

```python
total_destinations_ewr = destinations_ewr['dest'].nunique()

print('Total Destinations Newark:', total_destinations_ewr)
```

```python
destinations_ewr.info()
```

```python
destinations_ewr['time_hour'] = pd.to_datetime(destinations_ewr['time_hour'])
```

```python
destinations_ewr.info()
```

```python
destinations_ewr['date'] = destinations_ewr['time_hour'].dt.normalize()

destinations_ewr.head()
```

```python
destinations_by_day = destinations_ewr.groupby('day').agg({'dest': 'nunique'})

destinations_by_day.head()
```

```python
mean_destinations_by_day = destinations_by_day.mean()

print(round(mean_destinations_by_day, 0))
```

```python
destinations_by_month = destinations_ewr.groupby('month').agg({'flight': 'count'})

print(destinations_by_month.sort_values(by = 'flight', ascending = False))
```
